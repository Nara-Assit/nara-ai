```python
import numpy as np 
import pandas as pd
import os
```

```python
data1 = np.load('/kaggle/input/batch1/first100.npz')  # or 'data_arrays_compressed.npz'
X1 = data1['X']
y1 = data1['y']


data2 = np.load('/kaggle/input/second-and-third-batch/second100.npz')  # or 'data_arrays_compressed.npz'
X2 = data2['X']
y2 = data2['y']

data3 = np.load('/kaggle/input/second-and-third-batch/third100.npz')  # or 'data_arrays_compressed.npz'
X3 = data3['X']
y3 = data3['y']
```

```python
X = np.concatenate((X1,X2,X3))
y = np.concatenate((y1,y2,y3))
```

```python
X.shape , y.shape
```

```python
from sklearn.model_selection import train_test_split
# X is your features, y is your target labels

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.25, 
    random_state=42, 
    stratify=y 
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras import layers, regularizers

```

```python
actions = ["love" , "house", "mall" , "father" , "mosque" , "me" , "important", "happy" , "finish" , "thinking","baby" , "worry" , "normal" , "eat" , "stop" , "sad", "thanks" , "mother" , "hear" ,"good"]
```

```python
# detect and init the TPU
#tpu = tf.distribute.cluster_resolver.TPUClusterResolver()
#
## instantiate a distribution strategy
#tf.tpu.experimental.initialize_tpu_system(tpu)
#tpu_strategy = tf.distribute.TPUStrategy(tpu)
```

```python
model = Sequential()
model.add(LSTM(64, return_sequences=True, activation='leaky_relu', input_shape=(30,144)))
model.add(LSTM(128, return_sequences=True, activation='leaky_relu'))
model.add(LSTM(64, return_sequences=False, activation='leaky_relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(np.array(actions).shape[0], activation='softmax'))
```

```python
model.compile(optimizer="adam", loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train, num_classes=20)
y_test = to_categorical(y_test, num_classes=20)
```

```python
model.fit(X_train, y_train, epochs=50 , batch_size=32 , validation_data=(X_test ,y_test))
```
